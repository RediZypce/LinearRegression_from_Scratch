# Linear Regression from Scratch

Linear regression is a fundamental statistical and machine learning technique used for analyzing and modeling the relationship between a dependent variable (target) and one or more independent variables (features or predictors) by fitting a linear equation to observed data. It is a supervised learning algorithm, meaning it learns from labeled data where the relationship between the input variables and the target variable is known. Here's a detailed description of linear regression and why it is helpful:



## Description of Linear Regression

* Mathematical Formulation: Linear regression assumes that the relationship between the independent variables (X) and the dependent variable (Y) can be expressed as a linear combination of the independent variables.

* Goal: The goal of linear regression is to find the best-fitting linear model by estimating the values of $\beta$ coefficients that minimize the sum of squared errors (the difference between predicted and actual values).

* Ordinary Least Squares (OLS): Linear regression typically employs the OLS method, which minimizes the sum of the squared differences between predicted and actual values. This approach finds the optimal values for $\beta$ through mathematical optimization.


## Why Linear Regression is Helpful:

* Interpretability: Linear regression models are easy to interpret. The coefficients ($\beta$ values) provide insights into the relationship between each independent variable and the dependent variable. A positive coefficient indicates a positive correlation, and a negative coefficient indicates a negative correlation.

* Simplicity: Linear regression is a straightforward and simple algorithm, making it a good starting point for regression problems. It's easy to implement and understand, making it accessible to beginners.

* Useful for Baseline Models: Linear regression can serve as a baseline model for regression tasks. It provides a simple benchmark against which more complex models can be compared.

* Insight Generation: Linear regression can be used for feature selection and feature engineering. By analyzing the coefficients, you can identify which features have the most significant impact on the target variable.

* Assumptions Testing: Linear regression helps validate several assumptions about your data, such as linearity, independence of errors, constant variance, and normality of residuals. Detecting violations of these assumptions can guide data preprocessing and model selection.

* Predictive Power: Linear regression can provide reasonably accurate predictions when the relationship between the variables is approximately linear.

* Regression Analysis: It is widely used in fields like economics, finance, social sciences, and healthcare for analyzing and understanding the impact of independent variables on a dependent variable.

* Foundation for Advanced Techniques: Linear regression serves as the basis for more advanced regression techniques, such as ridge regression, lasso regression, and logistic regression.

```python
# Import libraries

import numpy as np
import matplotlib.pyplot as plt
```

In this example, we'll create a simple linear regression model to predict a person's weight based on their height. We'll implement gradient descent to optimize the model.

```python
# Generate some synthetic data for height and weight

np.random.seed(0)
height = np.random.randint(140, 190, 100)
weight = 0.5 * height + 30 + np.random.normal(0, 10, 100)
```

```python
# Normalize the data
height_normalized = (height - np.mean(height)) / np.std(height)
weight_normalized = (weight - np.mean(weight)) / np.std(weight)
```

```python
# Initialize the parameters (slope and intercept) randomly

slope = np.random.rand()
intercept = np.random.rand()
```

```python
# Set hyperparameters

learning_rate = 0.01
epochs = 1000
```

```python
# Lists to store the history of parameter updates and cost

slope_history = []
intercept_history = []
cost_history = []
```

```python
# Gradient Descent
for epoch in range(epochs):
    # Predicted values using the current slope and intercept
    predicted = slope * height_normalized + intercept
    
    # Calculate the cost (mean squared error)
    error = predicted - weight_normalized
    cost = np.mean(error ** 2)
    
    # Calculate the gradients
    slope_gradient = np.mean(error * height_normalized)
    intercept_gradient = np.mean(error)
    
    # Update the parameters using the gradients and learning rate
    slope -= learning_rate * slope_gradient
    intercept -= learning_rate * intercept_gradient
    
    # Store the history for visualization
    slope_history.append(slope)
    intercept_history.append(intercept)
    cost_history.append(cost)
```

```python
# Plot the data and regression line
plt.scatter(height_normalized, weight_normalized, label='Data Points')
plt.plot(height_normalized, slope * height_normalized + intercept, color='red', label='Regression Line')
plt.xlabel('Normalized Height')
plt.ylabel('Normalized Weight')
plt.legend()
plt.show()
```

```python
# Plot the cost function over iterations
plt.plot(range(epochs), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function over Iterations')
plt.show()
```

```python
# Visualize how gradient descent updated the parameters
plt.plot(range(epochs), slope_history, label='Slope')
plt.plot(range(epochs), intercept_history, label='Intercept')
plt.xlabel('Iterations')
plt.ylabel('Parameters')
plt.legend()
plt.title('Parameter Updates over Iterations')
plt.show()
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the cost function over parameter values
ax.plot(slope_history, intercept_history, cost_history, label='Cost Function')
ax.set_xlabel('Slope')
ax.set_ylabel('Intercept')
ax.set_zlabel('Cost')
ax.set_title('3D Visualization of Cost Function over Parameters')

# Plot the final parameter values
ax.scatter([slope], [intercept], [cost], color='red', label='Final Parameters')

# Show legend
ax.legend()

# Show the plot
plt.show()
```
